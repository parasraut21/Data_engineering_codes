# tuple duplication

```python
import pandas as pd

# Creating a hypothetical dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Bob'],
    'Age': [20, 22, 21, 20, 23, 22],
    'TestScore': [85, 90, 88, 85, 92, 90]
}

original_dataset = pd.DataFrame(data)
print("Original Dataset:")
print(original_dataset)
```

```python
# Step 3: Identify Duplicates
duplicate_rows = original_dataset[original_dataset.duplicated()]
print("\nDuplicate Rows:")
print(duplicate_rows)
```

```python
# Step 4: Remove Duplicates
clean_dataset = original_dataset.drop_duplicates()
print("\nCleaned Dataset:")
print(clean_dataset)
```

```python
# Step 5: Handling Duplicate Tuples
aggregated_dataset = original_dataset.groupby('Name').agg({'Age': 'mean', 'TestScore': 'mean'}).reset_index()
print("\nAggregated Dataset:")
print(aggregated_dataset)
```

```python






# Step 6: Save Processed Dataset
aggregated_dataset.to_csv('processed_dataset.csv', index=False)
print("\nProcessed Dataset saved to 'processed_dataset.csv'")
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data, index=['X', 'Y', 'Z'])

# Using iloc
print("Using iloc:")
print(df.iloc[0, 1])  # Access the element in the first row and second column
print()

# Using loc
print("Using loc:")
print(df.loc['X', 'B'])  # Access the element in the row labeled 'X' and column labeled 'B'
```
